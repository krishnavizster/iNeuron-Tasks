```python
with open('carevaluation.data', 'r') as f:
    carevaluation_data = csv.reader(f, delimiter = '\n')
    for i in carevaluation_data:
        row_data = i[0].split(',')
        print(row_data)
```

```python
#4 Dump all the data into car table 
with open('carevaluation.data', 'r') as f:
    carevaluation_data = csv.reader(f, delimiter = '\n')
    for i in carevaluation_data:
        row_data = tuple(i[0].split(','))
        query = "INSERT INTO cardataset.car2 VALUES {}".format(row_data)
        cursor.execute(query)
```
